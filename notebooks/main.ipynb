# Conclusion:

Cette étude démontre que le lien entre pollution de l'air et niveau de vie en Île-de-France ne suit pas une logique binaire, mais repose sur des mécanismes complexes de ségrégation spatiale. L'analyse révèle que l'exposition aux polluants ne touche pas uniquement les plus précaires : les centres urbains denses et riches présentent également des niveaux de pollution élevés.

Cependant, l'utilisation de méthodes statistiques a permis d'identifier quatre profils territoriaux types qui nuancent cette réalité :
- Une banlieue aisée et des zones rurales relativement préservées.
- Un hyper-centre riche qui fait le choix de la densité malgré la pollution.
- Un pôle de cumul de vulnérabilités, où précarité sociale et forte pollution se rejoignent, illustrant une situation flagrante d'injustice environnementale.

En conclusion, l'exposition à la pollution répond à des logiques différentes selon le territoire : elle est le résultat d'un arbitrage ou d'un privilège dans les zones aisées, tandis qu'elle relève d'une relégation subie pour les populations les plus fragiles.
